```python
import pandas as pd

# List of file names
file_names = [
    "JC-202305-citibike-tripdata.csv",
    "JC-202304-citibike-tripdata.csv",
    "JC-202303-citibike-tripdata.csv",
    "202305-citibike-tripdata.csv",
    "202304-citibike-tripdata.csv",
    "202303-citibike-tripdata.csv"
]

# Create an empty list to store dataframes
dfs = []

# Read and append each file into the list
for file_name in file_names:
    df = pd.read_csv(file_name)
    dfs.append(df)

# Combine all dataframes into one
combined_df = pd.concat(dfs)

# Optional: Reset the index of the combined dataframe
combined_df.reset_index(drop=True, inplace=True)

# Print the combined dataframe
combined_df.head()
```

```python
combined_df.info()
```

```python
import pandas as pd

# Assuming 'combined_df' is the original combined dataframe

# Create a copy of the original dataframe
df = combined_df.copy()

# Convert 'started_at' column to datetime
df['started_at'] = pd.to_datetime(df['started_at'])

# Define time periods
time_periods = {
    'morning': (6, 11, range(6, 12)),
    'afternoon': (12, 17, range(12, 18)),
    'evening': (18, 23, range(18, 24)),
    'overnight': (0, 5, range(0, 6))
}

# Function to assign time period based on the hour of 'started_at'
def assign_time_period(row):
    hour = row['started_at'].hour
    for period, (start, end, hour_range) in time_periods.items():
        if hour in hour_range:
            return period

# Add new column 'time_period' based on 'started_at'
df['time_period'] = df.apply(assign_time_period, axis=1)

# Print the updated dataframe
print(df)


```

```python
df_weekday=df.copy()

df_weekday['weekday']=df.groupby(['start_station_id']).agg(
    start_station_name=('start_station_name', 'first'),
    start_lat=('start_lat', 'first'),
    start_lng=('start_lng', 'first'),
    average_ride_time=('ride_time', 'mean'),
    riders=('ride_id', 'count')
).reset_index()
```

```python
# Convert 'ended_at' column to datetime
df['ended_at'] = pd.to_datetime(df['ended_at'])

# Calculate ride duration in minutes
df['ride_time'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60

# Convert ride_time to int
df['ride_time'] = df['ride_time'].astype(int)

# Print the updated dataframe
df.head()
```

```python
# Extracting date from 'started_at' column
df_timeline=df.copy()

# Extracting date from 'started_at' column
df_timeline['date'] = df_timeline['started_at'].dt.date

# Grouping by date and time_period
grouped_df = df_timeline.groupby(['date', 'time_period'])

# Calculating the average of ride_time and count of ride_id
result_df = grouped_df.agg({'ride_time': 'mean', 'ride_id': 'count'})

# Resetting the index
result_df = result_df.reset_index()

# Renaming the columns
result_df = result_df.rename(columns={'ride_time': 'Average Ride Time', 'ride_id': 'Total Rides'})

# Reordering the columns
result_df = result_df[['date', 'time_period', 'Total Rides', 'Average Ride Time']]

# Define the file path for the output CSV file
output_file_path = 'Output/df_timeline.csv'

# Export the dataframe to a CSV file using pandas
result_df.to_csv(output_file_path, index=False)

result_df.head()
```

```python
df=df[(df['ride_time']>=0) & (df['ride_time']<=1440)]

df.info()
```

```python
# Grouping by day of week and time_period
grouped_df = df.groupby([df['started_at'].dt.dayofweek, 'time_period'])

# Calculating sum of ride_id and average of ride_time
result_df = grouped_df.agg({'ride_id': 'count', 'ride_time': 'mean'})

# Renaming the columns
result_df = result_df.rename(columns={'ride_id': 'Total Rides', 'ride_time': 'Average Ride Time'})

# Resetting the index
result_df = result_df.reset_index()

# Mapping day of week to actual weekday names
result_df['Day of Week'] = result_df['started_at'].map({0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'})

# Dropping the day of week numeric column
result_df = result_df.drop('started_at', axis=1)

# Reordering the columns
result_df = result_df[['Day of Week', 'time_period', 'Total Rides', 'Average Ride Time']]

# Define the file path for the output CSV file
output_file_path = 'Output/weekday_df.csv'

# Export the dataframe to a CSV file using pandas
result_df.to_csv(output_file_path, index=False)

result_df.head()
```

```python
grouped_start_df = df.groupby(['start_station_id', 'time_period']).agg(
    start_station_name=('start_station_name', 'first'),
    start_lat=('start_lat', 'first'),
    start_lng=('start_lng', 'first'),
    average_ride_time=('ride_time', 'mean'),
    riders=('ride_id', 'count')
).reset_index()

# Add the 'daily_riders' column
grouped_start_df['daily_riders'] = grouped_start_df['riders'] / 90

updated_start_df = grouped_start_df.copy()

# Group the dataframe by 'start_station_id' and update 'start_lat' and 'start_lng' to the first occurrence within each group
updated_start_df['start_lat'] = grouped_start_df.groupby('start_station_id')['start_lat'].transform(lambda x: x.iloc[0])
updated_start_df['start_lng'] = grouped_start_df.groupby('start_station_id')['start_lng'].transform(lambda x: x.iloc[0])

# Define the file path for the output CSV file
output_file_path = 'Output/updated_start_df.csv'

# Export the dataframe to a CSV file using pandas
updated_start_df.to_csv(output_file_path, index=False)

updated_start_df.head(10)
```
